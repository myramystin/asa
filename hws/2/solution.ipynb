```python

import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df = pd.read_excel('data0110.xlsx')
```

```python
df.head()
```

Предскажем столбец Средняя заработная плата научных работников (без внешних совместителей и работающих по договорам ГПХ)

```python
df.columns.values.tolist()
```

```python
y = df[['Средняя заработная плата научных работников (без внешних совместителей и работающих по договорам ГПХ)']].fillna(0)
```

Понятно, что много столбцов в датасете нерелевантные. попробуем выбрать самые релевантные: мне кажется, это будет регион, средний балл егэ, а также доходы вуза

некоторые признаки (регион например) придется закодировать числами)


```python
columns = [
    'Регион',
    'Средний балл ЕГЭ студентов, принятых по результатам ЕГЭ на обучение по очной форме по программам бакалавриата и специалитета за счет средств соответствующих бюджетов бюджетной системы РФ',
    'Доходы вуза от образовательной деятельности из иностранных источников, тыс. руб.',
    'Общий объем средств, поступивших (за отчетный год) от выполнения работ, услуг, связанных с научными, научно-техническими, творческими услугами и разработками, выполненных собственными силами'
]
```

```python
def solve(selected_columns):
    X = df[selected_columns].fillna(0)
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

    model = LinearRegression()
    model.fit(X_train, y_train)

    y_pred = model.predict(X_test)

    mse_error = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)

    print("Среднеквадратическая ошибка:", mse_error)
    print("Коэффициент детерминации:", r2)
```

```python
le = LabelEncoder()
le.fit(df['Регион'])
df['Регион'] = le.transform(df['Регион'])
```

```python
solve(columns)
```
